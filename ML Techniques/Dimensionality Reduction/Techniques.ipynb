#### PCA (Principal Component Analysis)

```python
import pandas as pd
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt 

data = pd.read_csv("Iris.csv", index_col=0)

X = data.drop(columns=["Species"])
y = data["Species"].astype("category").cat.codes

plt.scatter(X["SepalLengthCm"], X["SepalWidthCm"], c=y)
plt.title("Previous")
plt.show()

pca = PCA(n_components=2)
x_transformed = pca.fit_transform(X)
# print(x_transformed)

plt.scatter(x_transformed[:,0], x_transformed[:,1], c=y)
plt.title("After")
plt.show()
```

```python
print("Model Components", pca.components_)
print("Amount variance that the components captured: ", pca.explained_variance_)
print("Explained Variance Ratio", pca.explained_variance_ratio_)
```

```python
import numpy as np
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

print("Cumulative variance", cumulative_variance)

# The pca is good as it captures around 98% of variance with just 2components which is greater than 95%
```

#### LDA (Linear Discriminant Analysis)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

plt.scatter(X["SepalLengthCm"], X["SepalWidthCm"], c=y)
plt.title("Previous")
plt.show()

lda = LinearDiscriminantAnalysis(n_components=2)
x_transformed = lda.fit_transform(X,y)
# print(x_transformed)

plt.scatter(x_transformed[:,0], x_transformed[:,1], c=y)
plt.title("After")
plt.show()

print(lda.explained_variance_ratio_)
print(np.cumsum(lda.explained_variance_ratio_))
```
