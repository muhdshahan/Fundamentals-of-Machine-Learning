#### KNN approach

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

songs = pd.read_csv("songData.csv", index_col=0)
print("Songs Dataframe\n",songs.head())

table = songs.pivot_table(index="'userID'", columns="'songID'", values="'rating'", fill_value=0)
print("Pivoted Table\n",table.tail())
```

```python
from sklearn.neighbors import NearestNeighbors

# automatically computes cosine similarity so need to do seperately if want to do seperately then its a diff method
knn = NearestNeighbors(metric="cosine", algorithm="brute")
knn.fit(table)
distance, neighbors = knn.kneighbors(table, n_neighbors=5)
neighbors
```

```python
# lets find the recomendation

def similar_users(user_id, similarity_matrix, top_n=4):
    return list(similarity_matrix[user_id][:top_n])
    
similar_users(500,neighbors)
```
